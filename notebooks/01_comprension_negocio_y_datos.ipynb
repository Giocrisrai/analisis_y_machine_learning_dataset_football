```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", context="notebook")
fig, axes = plt.subplots(1, 3, figsize=(10, 3), sharey=True)
for ax, col in zip(axes, b365, strict=True):
    sns.histplot(match[col].dropna(), ax=ax, kde=True, stat="density")
    ax.set_title(col)
plt.suptitle("Distribución de cuotas B365 (con datos)")
plt.tight_layout()
plt.show()
```
